#### d) Write your own bagging method

```python
from sklearn.utils import resample
```

```python
###Define a new base decision tree classifier
 
Base_DecisionTree = DecisionTreeClassifier(random_state=2)
```

```python
results = []

for k in range(0, 19):
    
    #1) Make a bootstrap sample of the original dataset
    bootstrap_size = 0.8*len(X_train)
    #2) Command to generate a random bootstrap dataset
    X_train_sample = resample(X_train, n_samples=bootstrap_size, random_state=k,replace = True)
    y_train_sample = resample(y_train, n_samples=bootstrap_size, random_state=k,replace = True)
    #3) Train the new base decision tree classifier
    Base_DecisionTree.fit(X_train_sample,y_train_sample)
    #4) Test this "base classifier" on the original testing dataset and save the predictions 
    y_predict_bdt = Base_DecisionTree.predict(X_test)
    print(y_predict_bdt)   
    results.append(y_predict_bdt)
    
#5) Save the result into a matrix    
print(results) 
mat = np.matrix(results)
#6) Print the matrix and size of a matrix
print(mat)
mat.shape
#7) Transpose the matrix so that we get a (45,19) matrix 
mat2 = mat.transpose()
#8) Print the matrix and size of a the new matrix
print(mat2)
mat2.shape
```

```python
#9)Take the sum of each row and store it in an array
sum_of_col = np.sum(mat2, axis=1)
print (sum_of_col)
#10)Store the length of the array in a variable
length=len(sum_of_col)
```

```python
#11)Using the length of the array initialize a for loop 
res=[]
#12)Initialize a for loop to check if the sum of each column is lesser than 9
#if it is lesser than 9 then assign the value of 0 to a variable
#append these values into another list
for i in range(0,length):
    if sum_of_col[i] < 9:
                         m = 0
    else:
        m=1
    res.append(m)  
print(res)
```

```python
#14) Check the accuracy for our bagging procedure
score_bag = accuracy_score(y_test, res)
print(score_bag)
```

###### Accuracy for our bagging method is : 0.9111
